```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
n_evecs = 20
```

```python
indices = jnp.arange(0,n_evecs,1)[::-1]
```

```python
evecs_soa = [jnp.load(f"/Users/eartsu/new_model/testing/nm//bits_of_data/hessian_evecs_etc/shelf/new/ts_evec_soa_2km_{i}.npy").flatten() for i in indices]
evecs_ad  = [jnp.load(f"/Users/eartsu/new_model/testing/nm//bits_of_data/hessian_evecs_etc/shelf/new/ts_evec_ad_2km_{i}.npy").flatten()  for i in indices]

# evecs_soa = [jnp.load(f"/Users/eartsu/new_model/testing/nm//bits_of_data/hessian_evecs_etc/stream/new/ts_evec_soa_1km_{i}.npy").flatten() for i in indices]
# evecs_ad  = [jnp.load(f"/Users/eartsu/new_model/testing/nm//bits_of_data/hessian_evecs_etc/stream/new/ts_evec_ad_1km_{i}.npy").flatten()  for i in indices]
```

```python
evec_mats_soa = jnp.column_stack(evecs_soa)
evec_mats_ad  = jnp.column_stack(evecs_ad)
```

```python
def principal_angles(A, B):
    """
    Given two matrices A, B with the same number of rows and k columns,
    compute principal angles between the column spaces of A and B.
    Returns angles in radians (sorted ascending).
    """
    # Orthonormalise columns (QR)
    QA, _ = jnp.linalg.qr(A)  # QA: (n, k)
    QB, _ = jnp.linalg.qr(B)  # QB: (n, k)
    
    # SVD of QA^T QB -> singular values are cos(theta_i)
    M = QA.T @ QB
    s = jnp.linalg.svd(M, compute_uv=False)
    
    # Numerical safety
    s = jnp.clip(s, 0.0, 1.0)
    thetas = jnp.arccos(s)  # radians
    return jnp.sort(thetas)  # ascending
```

```python
theta_max = []
theta_min = []
theta_mean = []
grassmann_geodesic = []  # ||theta||_2 on the Grassmann manifold
chordal = []             # sqrt(sum(sin(theta)^2))
projection_dist = []     # sin(theta_max) = ||P_A - P_B||_2
all_thetas = []

for k in range(1,n_evecs):
    thetas = principal_angles(evec_mats_soa[:, :k], evec_mats_ad[:, :k])
    all_thetas.append(thetas)
    theta_max.append(thetas[-1])
    theta_mean.append(thetas.mean())
    theta_min.append(thetas[0])
    grassmann_geodesic.append(jnp.linalg.norm(thetas))
    sines = jnp.sin(thetas)
    chordal.append(jnp.linalg.norm(sines))
    projection_dist.append(sines[-1])  # equals sin(max angle)
```

```python
plt.scatter(range(1,n_evecs), projection_dist)
plt.show()
```

```python
plt.scatter(range(1,n_evecs), jnp.degrees(jnp.array(theta_min)))
plt.show()
```

```python
plt.scatter(range(1,n_evecs), chordal)
plt.show()

plt.scatter(range(1,n_evecs), grassmann_geodesic)
plt.show()
```

```python
plt.scatter(range(1,n_evecs), jnp.degrees(jnp.array(theta_min)))
plt.show()
```

```python
max_angles = jnp.degrees(jnp.array(theta_max))
mean_angles = jnp.degrees(jnp.array(theta_mean))
```

```python
plt.figure(figsize=(7,5))
plt.plot(range(1,n_evecs), max_angles, 'o-', label='max principal angle')
plt.plot(range(1,n_evecs), mean_angles, 'o-', label='mean principal angle')
plt.xlabel('Number of eigenvectors included (k)')
plt.ylabel('Angle (degrees)')
plt.title('Subspace divergence between AD and SOA eigenvector sets')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
kernel = jnp.array([1, 1, 1, 1, 1]) / 3.0

smoothed_max_angles = jnp.convolve(max_angles, kernel, mode='same')
smoothed_mean_angles = jnp.convolve(mean_angles, kernel, mode='same')
```

```python
plt.figure(figsize=(7,5))
plt.plot(range(1,n_evecs), smoothed_max_angles, 'o-', label='max principal angle')
plt.plot(range(1,n_evecs), smoothed_mean_angles, 'o-', label='mean principal angle')
plt.xlabel('Number of eigenvectors included (k)')
plt.ylabel('Angle (degrees)')
plt.title('Subspace divergence between AD and SOA eigenvector sets')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
